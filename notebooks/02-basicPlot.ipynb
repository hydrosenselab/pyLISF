```python
# Plot 

def mapPlot(style,cmap,boundaries,variable,norm):
    plt.style.use('{}'.format(style))
    plt.figure(figsize=(8,11))
    cmap = colors.ListedColormap(cmap)
    if norm:
        normSet = colors.BoundaryNorm(boundaries, cmap.N, clip=True)
    im=xr.plot.imshow(dat['{}'.format(variable)].isel(time=87),norm=normSet,cmap = cmap,add_labels=False, add_colorbar=False)
    cb = plt.colorbar(im, orientation="horizontal", pad=0.05)
    cb.set_label(r'Streamflow ($m^{3}$/$s$)', labelpad=-70, y =0.5) #To-do
    plt.grid(False)
    plt.axis('off')
    plt.title('\nIndia Water Model \n' + str(dat['time'].values)[2:13]+'\n',
              loc='left',fontdict= {'weight': 'bold','size':22})
    plt.show()
```

```python
mapPlot('fivethirtyeight',['#f7f7f7','#2166ac','#4393c3','#abdda4','#e6f598','#fee08b','#fdae61','#f46d43','#d53e4f','#9e0142'],[0,10,50,100,250,500,1000,2500,5000,10000,20000], 'Streamflow_tavg', True  )
```
